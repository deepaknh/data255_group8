```python
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import contractions
import spacy

# Download nltk resources
nltk.download('stopwords')
nltk.download('punkt')

# Initialize spaCy's English model
nlp = spacy.load('en_core_web_sm')

# Initialize stopwords
stop_words = set(stopwords.words('english'))

def clean_twitter_data(df):
    cleaned_data = []

    for post in df['posts']:
        # Remove URLs
        post = re.sub(r'https?://\S+|www\.\S+', '', post)

        # Expand Contractions
        post = contractions.fix(post)

        # Convert text to lowercase
        post = post.lower()

        # Tokenize the post
        tokens = word_tokenize(post)

        # Lemmatize tokens using spaCy
        lemmatized_tokens = [token.lemma_ for token in nlp(' '.join(tokens))]

        # Remove stopwords and non-alphanumeric tokens
        filtered_tokens = [word for word in lemmatized_tokens if word.isalnum() and word not in stop_words]

        # Reconstruct the post
        post = ' '.join(filtered_tokens)

        cleaned_data.append(post)

    return pd.DataFrame(cleaned_data, columns=['posts'])

# Clean the dataset
cleaned_df = clean_twitter_data(df)

# Save the cleaned dataset
cleaned_df.to_csv('mbti_labelled_cleaned.csv', index=False)
```
